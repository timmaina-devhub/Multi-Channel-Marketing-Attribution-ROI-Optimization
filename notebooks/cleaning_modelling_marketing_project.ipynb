**Attribution Modelling....**

**A) `Markov Chain Attribution Model`**

**Building proper paths**

```python
def build_path(row):
    path = ["START"] + row['journey_path']
    
    if row['converted'] == 1:
        path.append("CONVERSION")
    else:
        path.append("NULL")
    
    return path

customers_journey_path['path'] = customers_journey_path.apply(build_path, axis=1)
```

```python
customers_journey_path.head()
```

```python
customers_journey_path['path'].iloc[1]
```

**Creating transition counts**

```python
from collections import defaultdict

transition_counts = defaultdict(int)

for path in customers_journey_path['path']:
    for i in range(len(path) - 1):
        transition_counts[(path[i], path[i+1])] += 1
```

**Converting transition counts to transition probabilities**

```python
df_transitions = pd.DataFrame([
    {'from': k[0], 'to': k[1], 'count': v}
    for k, v in transition_counts.items()
])

df_transitions['prob'] = df_transitions.groupby('from')['count'].transform(lambda x: x / x.sum())

df_transitions
```

**Building transition matrix**

```python
states = list(set(df_transitions['from']).union(set(df_transitions['to'])))

matrix = pd.DataFrame(0, index=states, columns=states)

for _, row in df_transitions.iterrows():
    matrix.loc[row['from'], row['to']] = row['prob']
```

**Computing Baseline Conversion Probability (Simulation)**

```python
def simulate(matrix, n_sim=10000):
    conversions = 0
    
    for _ in range(n_sim):
        current = "START"
        
        while current not in ["CONVERSION", "NULL"]:
            probs = matrix.loc[current]
            
            if probs.sum() == 0:
                break
            
            next_state = random.choices(probs.index, weights=probs.values)[0]
            current = next_state
        
        if current == "CONVERSION":
            conversions += 1
    
    return conversions / n_sim

baseline_conversion_rate = simulate(matrix)
```

```python
baseline_conversion_rate
```

**Removal Effect (Core Attribution Logic)**

```python
def remove_channel(matrix, channel):
    new_matrix = matrix.copy()
    
    # Remove channel
    new_matrix.loc[channel, :] = 0
    new_matrix.loc[:, channel] = 0
    
    # Re-normalize probabilities
    new_matrix = new_matrix.div(new_matrix.sum(axis=1), axis=0).fillna(0)
    
    return new_matrix
```

**Computing removal effects:**

```python
removal_effects = {}

for channel in matrix.index:
    if channel in ["START", "CONVERSION", "NULL"]:
        continue
    
    modified_matrix = remove_channel(matrix, channel)
    new_conversion_rate = simulate(modified_matrix)
    
    effect = (baseline_conversion_rate - new_conversion_rate) / baseline_conversion_rate
    removal_effects[channel] = effect
```

```python
removal_effects
```

**Attribution Scores**

**Comverting all dicts to a dataFrame**

```python
df_results=pd.DataFrame({
    'channel': list(removal_effects.keys()),
    'removal_effect': list(removal_effects.values()),
})
```

```python
df_results['total_effect']=sum(removal_effects.values())
```

```python
df_results['attribution_fraction']= (df_results['removal_effect'] / df_results['total_effect'])

df_results['attribution_percentage']=(df_results['attribution_fraction'] * 100)
```

```python
df_results=df_results.sort_values(by='attribution_percentage',ascending=False).reset_index()
df_results=df_results.drop(columns=['total_effect','index'])
df_results
```
